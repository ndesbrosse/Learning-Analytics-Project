```python
df_final=pd.merge(mdl_grade_grades, mdl_grade_items, left_on="itemid", right_on="id", how="inner")
```

```python
df_final=df_final[df_final["itemmodule"]=='assign']
```

```python
df_final = df_final.dropna(subset=['finalgrade'])
```

```python
display(mdl_grade_grades[(mdl_grade_grades['userid'] == 113924) & (mdl_grade_grades['itemid'] == 13553)])
```

```python
df_final["finalgrade"] = pd.to_numeric(df_final["finalgrade"])
```

```python
df_final["grade"]=round(((df_final["finalgrade"]-df_final["rawgrademin"])/(df_final["rawgrademax"]-df_final["rawgrademin"]))*20,2)
```

```python
display(df_final)
```

```python
display(df_final[(df_final['courseid'] == 45757) & (df_final['itemid'] == 228813)])
```

```python
df_model = df_final[["itemid", "userid","courseid","rawgrademax","aggregationweight","grade"]]
```

```python
df_model["aggregationweight"] = pd.to_numeric(df_model["aggregationweight"],errors="coerce")

# Calcul de la moyenne des grades groupée par userid
average_grade_by_userid = round(df_model.groupby('userid')['grade'].mean().reset_index(),2)

# Renommer les colonnes pour plus de clarté
average_grade_by_userid.columns = ['userid', 'user_grade_avg']


# Calcul de la moyenne des grades groupée par userid
average_grade_by_courseid = round(df_model.groupby('courseid')['grade'].mean().reset_index(),2)

# Renommer les colonnes pour plus de clarté
average_grade_by_courseid.columns = ['courseid', 'course_grade_avg']


df_model=pd.merge(df_model, average_grade_by_userid, left_on="userid", right_on="userid", how="left")

df_model=pd.merge(df_model, average_grade_by_courseid, left_on="courseid", right_on="courseid", how="left")

df_model['course_user_grade_avg'] = round(df_model.groupby(['courseid', 'userid'])['grade'].transform('mean'),2)
```

```python
display(df_model)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV

# Préparation des données
X = df_model.drop(columns=['grade'])  # Toutes les colonnes sauf 'grade'
y = df_model['grade']  # La colonne cible

# Diviser les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Création et entraînement du modèle
model = RandomForestRegressor(random_state=42,max_depth=20, n_estimators=200,min_samples_leaf=2,min_samples_split=10)

model.fit(X_train, y_train)

# Faire des prédictions
y_pred = model.predict(X_test)

y_pred=[round(x, 2) for x in y_pred]

# Créer un DataFrame pour comparer les valeurs réelles et les prédictions
comparison_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})

# Afficher le tableau comparatif
display(comparison_df)

# Évaluation du modèle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R^2 Score: {r2}")

# Afficher l'importance des caractéristiques
importances = model.feature_importances_
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
# Création du graphique
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['Feature'], feature_importance['Importance'], align='center')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance for Grade Prediction')
plt.gca().invert_yaxis()  # Inverse l'axe Y pour une meilleure lisibilité
plt.show()
```

```python
# Créer le graphique
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.6, edgecolor='k')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Perfect Prediction')
plt.xlabel('Valeurs Réelles (y_test)')
plt.ylabel('Valeurs Prédites (y_pred)')
plt.title('Comparaison des Valeurs Réelles et Prédites')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
import joblib

# Préparation des données
X = df_model.drop(columns=['grade'])  # Toutes les colonnes sauf 'grade'
y = df_model['grade']  # La colonne cible

# Diviser les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Définir le modèle
xgb_model = XGBRegressor(random_state=42,learning_rate=0.1,max_depth=7,n_estimators=200,subsample=1.0)


xgb_model.fit(X_train, y_train)

# Faire des prédictions
y_pred = xgb_model.predict(X_test)

joblib.dump(xgb_model, 'modele_notes.pkl')


y_pred=[round(x, 2) for x in y_pred]

# Créer un DataFrame pour comparer les valeurs réelles et les prédictions
comparison_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})

# Afficher le tableau comparatif
display(comparison_df)

# Évaluation du modèle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R^2 Score: {r2}")

# Afficher l'importance des caractéristiques
importances = xgb_model.feature_importances_
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
# Création du graphique
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['Feature'], feature_importance['Importance'], align='center')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance for Grade Prediction')
plt.gca().invert_yaxis()  # Inverse l'axe Y pour une meilleure lisibilité
plt.show()
```

```python
# Définir les intervalles et les classes
bins = [0, 10, 20]  # Les bornes des groupes
labels = [0, 1]  # Les classes correspondantes

df_model_class=df_model.copy()

# Créer une nouvelle colonne 'grade_class' pour les classes
df_model_class['grade_class'] = pd.cut(df_model_class['grade'], bins=bins, labels=labels, include_lowest=True)

# Vérifier les résultats
print(df_model_class[['grade', 'grade_class']].head(40))
```

```python
df_model_class=df_model_class.drop(columns=['grade']) 
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Assurez-vous que la colonne 'grade_class' a été créée
# Transformation de 'grade_class' en valeurs numériques si nécessaire
df_model_class['grade_class'] = df_model_class['grade_class'].astype(int)

# Séparation des caractéristiques (features) et de la cible (target)
X = df_model_class.drop(columns=['grade_class'])  # Caractéristiques
y = df_model_class['grade_class']  # Classe cible

# Division des données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42,stratify=y)

# Création et entraînement du modèle
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

joblib.dump(model, 'modele_classe_notes.pkl')

# Évaluation du modèle
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))




# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)

# Visualisation
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.title('Confusion Matrix')
plt.show()
```
